```python
def multi_coil_NUFFT_forward(NufftObj,  images):
    [ny, nx, nc] = images.shape
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    data = np.zeros([nl, nc], dtype = complex)
    for c in range(nc): 
        data[:,c] = NufftObj.forward(images[:,:,c])
    return data
def multi_coil_NUFFT_adjoint(NufftObj,  data):
    [ny, nx] = NufftObj.st["Nd"]
    nc = data.shape[-1]
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    images = np.zeros([ny, nx, nc], dtype = complex)
    for c in range(nc): 
        images[:,:,c] = NufftObj.solve(data[:,c], solver='cg', maxiter= 100)
    return images
```

```python
data = multi_coil_NUFFT_forward(NufftObj,  images)
selfadjoint = multi_coil_NUFFT_adjoint(NufftObj,  data)
show(rsos(selfadjoint))
```

```python
def espirit_coilmap(data, k = 6, r = 32, t = 0.01, c = 0.99):
    #   k: Parameter that determines the k-space kernel size. If X has dimensions (1, 256, 256, 8), then the kernel 
    #      will have dimensions (1, k, k, 8)
    #   r: Parameter that determines the calibration region size. If X has dimensions (1, 256, 256, 8), then the 
    #      calibration region will have dimensions (1, r, r, 8)
    #   t: Parameter that determines the rank of the auto-calibration matrix (A). Singular values below t times the
    #      largest singular value are set to zero.
    #   c: Crop threshold that determines eigenvalues "=1".
    [ny, nx, nc] = data.shape
    data = fft2c(data)
    data = data.reshape(ny, nx, 1, nc)
    images = ifft2c(data)
    esp = espirit(data, k, r, t, c)
    ip, proj, null = espirit_proj(images, esp)
    return np.squeeze(esp)[:,:,:,0]
```

```python
def PLOARKS_C(data, kx, ky):
    [ny, nx, nc] = data.shape
    data = np.moveaxis(data, -1, 0)
    mat = np.zeros([(ny-ky+1)*(nx-kx+1), kx * ky * nc], dtype = complex)
    idx = 0
    for y in range(max(1, ny - ky + 1)):
        for x in range(max(1, nx - kx + 1)):
            mat[idx, :] = data[:,y:y+ky, x:x+kx].reshape(1,-1)
            idx += 1
    return mat

def PLOARKS_Cinv(data, kx, ky, shape):
    [ny, nx, nc] = shape
    [nt, ks] = data.shape
    data = data.reshape(nt, nc, ky , kx)
    mat = np.zeros([nc, ny, nx], dtype = complex)
    count = np.zeros([nc, ny, nx])
    idx = 0
    for y in range(max(1, ny - ky + 1)):
        for x in range(max(1, nx - kx + 1)):
            mat[:,y:y+ky, x:x+kx] += data[idx]
            count[:,y:y+ky, x:x+kx] += 1 
            idx += 1
    mat = mat/count
    return np.moveaxis(mat, 0, -1) 
```

```python
def inspect_rank(data, thres = 0.01):
    U, S, VT = np.linalg.svd(data,full_matrices=False)
    s = np.copy(S)
    s = s / np.max(s)
    index = [ n for n,i in enumerate(s) if i > thres ][-1]
    yval = s[index] 
    S = np.diag(S)/np.max(np.diag(S))
    
    print("rank:"+str(index))
    
    S = np.diag(S)
    plt.figure(figsize=(8, 6), dpi=80)
    plt.subplot(2,1,1)
    plt.plot(S)
    plt.axhline(y =yval, color = 'r', linestyle = '-')
    plt.title('Singular Values')
    plt.subplot(2,1,2)
    plt.imshow(np.abs(np.transpose(VT)),aspect='auto', cmap = "gray")
    plt.show()
```

```python
def cal_rank(data, thres = 0.01):
    U, S, VT = np.linalg.svd(data,full_matrices=False)
    s = np.copy(S)
    s = s / np.max(s)
    index = [ n for n,i in enumerate(s) if i > 0.05 ][-1]
    return index 
```

```python
def l2(a, b):
    return np.linalg.norm(np.abs(a.flatten()) - np.abs(np.abs(b.flatten())))
```

```python
def multi_coil_NUFFT_forward(NufftObj,  images, sensMap):
    [ny, nx, nc] = sensMap.shape
    images = np.repeat(images[:, :,  np.newaxis], nc, axis=-1)
    images = images * sensMap 
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    data = np.zeros([nl, nc], dtype = complex)
    for c in range(nc): 
        data[:,c] = NufftObj.forward(images[:,:,c])
    return data
def multi_coil_NUFFT_adjoint(NufftObj,  data, sensMap):
    [ny, nx, nc] = sensMap.shape
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    images = np.zeros([ny, nx, nc], dtype = complex)
    for c in range(nc): 
        images[:,:,c] = NufftObj.solve(data[:,c], solver='dc', maxiter= 50)
    image = np.sum(images * np.conj(sensMap), -1)
    return image
def cg_sense(NufftObj, dataR, sensMap, nTR, nRO, ndim, maxit = 5, tol = 1e-6, lambd = 0):
    
    [ny, nx, nc] = sensMap.shape
    input_data = np.copy(dataR).reshape(nTR, nRO, nc)
    traj = NufftObj.st["om"].reshape(nTR, nRO, ndim)
    B = multi_coil_NUFFT_adjoint(NufftObj,  dataR.reshape(nTR * nRO, nc), sensMap)
    B = B.flatten()
    x = 0*B
    r = B 
    d = r 
    for ii in range(maxit):
        tmp = multi_coil_NUFFT_forward(NufftObj, d.reshape(ny, nx), sensMap)
        tmp = tmp.reshape(nTR, nRO, nc)
        tmp[:,:gap,:] = input_data[:,:gap,:]
        tmp = tmp.reshape(nTR*nRO, nc)
        Ad = multi_coil_NUFFT_adjoint(NufftObj,  tmp, sensMap) 
        Ad = Ad.flatten()
        a = np.dot(r,r)/(np.dot(d,Ad))
        x = x + a*d
        plt.figure()
        plt.imshow(np.abs(x.reshape(ny, nx)), cmap = "gray")
        plt.show()
        rn = r - a*Ad
        beta = np.dot(rn,rn)/np.dot(r,r)
        r=rn
        d = r + beta*d
    return x.reshape([ny, nx])
```

```python
from numpy import linalg 
def rank_approx(data, rank):
    U, S, VT = np.linalg.svd(data,full_matrices=False)
    return U[:,:rank] @ np.diag(S)[:rank, :rank] @  VT[:rank]

def multi_coil_NUFFT_forward(NufftObj,  images):
    [ny, nx, nc] = images.shape
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    data = np.zeros([nl, nc], dtype = complex)
    for c in range(nc): 
        data[:,c] = NufftObj.forward(images[:,:,c])
    return data
def multi_coil_NUFFT_adjoint(NufftObj,  data):
    [ny, nx] = NufftObj.st["Nd"]
    nc = data.shape[-1]
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    images = np.zeros([ny, nx, nc], dtype = complex)
    for c in range(nc): 
        images[:,:,c] = NufftObj.solve(data[:,c], solver='cg', maxiter= 100)
    return images
def low_rank_projection(NufftObj, dataR):
    kx = 7
    ky = 7
    thres = 0.01
    rank = 100
    traj = trajectory.reshape(nTR,nRO,2)
    dataR[:,0:gap,:] = 0
    log = []
    zte_data = np.copy(dataR).reshape(nTR*nRO, -1)
    for i in tqdm(range(50)):
        recon2 = multi_coil_NUFFT_adjoint(NufftObj, zte_data)
        recon2 = fft2c(recon2)
        C = PLOARKS_C(recon2, kx, ky)
        # inspect_rank(C, thres)
        Cr = rank_approx(C, rank)
        kc = PLOARKS_Cinv(Cr, kx, ky, recon2.shape)
        images = ifft2c(kc)
        zte_data = multi_coil_NUFFT_forward(NufftObj,  images)
        zte_data = zte_data.reshape(nTR, nRO, -1)
        zte_data[:,gap:,:] = dataR[:,gap:,:]
        log.append(l2(unsampled, zte_data[:,0:gap,:] ))
        zte_data = zte_data.reshape(nTR*nRO, -1)
    return zte_data.reshape(nTR* nRO, -1)
def multi_coil_NUFFT_forward_sens(NufftObj,  images, sensMap):
    [ny, nx, nc] = sensMap.shape
    images = np.repeat(images[:, :,  np.newaxis], nc, axis=-1)
    images = images * sensMap 
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    data = np.zeros([nl, nc], dtype = complex)
    for c in range(nc): 
        data[:,c] = NufftObj.forward(images[:,:,c])
    return data
def multi_coil_NUFFT_adjoint_sens(NufftObj,  data, sensMap):
    [ny, nx, nc] = sensMap.shape
    trajectory = NufftObj.st["om"]
    nl = trajectory.shape[0]
    images = np.zeros([ny, nx, nc], dtype = complex)
    for c in range(nc): 
        images[:,:,c] = NufftObj.solve(data[:,c], solver='dc', maxiter= 50)
    image = np.sum(images * np.conj(sensMap), -1)
    return image    
def cg_sense(NufftObj, dataR, sensMap, nTR, nRO, ndim, maxit = 5, tol = 1e-6, lambd = 0):
    
    [ny, nx, nc] = sensMap.shape
    input_data = np.copy(dataR).reshape(nTR, nRO, nc)
    traj = NufftObj.st["om"].reshape(nTR, nRO, ndim)
    B = multi_coil_NUFFT_adjoint_sens(NufftObj,  dataR.reshape(nTR * nRO, nc), sensMap)
    B = B.flatten()
    x = 0*B
    r = B 
    d = r 
    for ii in range(maxit):
        tmp = multi_coil_NUFFT_forward_sens(NufftObj, d.reshape(ny, nx), sensMap)
        tmp = tmp.reshape(nTR, nRO, nc)
        tmp[:,:gap,:] = input_data[:,:gap,:]
        tmp = tmp.reshape(nTR*nRO, nc)
        Ad = multi_coil_NUFFT_adjoint_sens(NufftObj,  tmp, sensMap) 
        Ad = Ad.flatten()
        a = np.dot(r,r)/(np.dot(d,Ad))
        x = x + a*d
        plt.figure()
        plt.imshow(np.abs(x.reshape(ny, nx)), cmap = "gray")
        plt.show()
        rn = r - a*Ad
        beta = np.dot(rn,rn)/np.dot(r,r)
        r=rn
        d = r + beta*d
    return tmp.reshape(nTR, nRO, nc)
```
